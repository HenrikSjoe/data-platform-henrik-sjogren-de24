  d) Find out the 10 oldest atheletes, their age and the sport


```python
duckdb.query("SELECT NAME, AGE, SPORT FROM df_os ORDER BY AGE DESC LIMIT 10").df()
```

e) Find out the 10 youngest atheletes, their age and the sport

```python
duckdb.query("SELECT NAME, AGE, SPORT, YEAR, MEDAL FROM df_os ORDER BY AGE LIMIT 10").df()
```

  e) Find out the five sports with highest median age

```python
duckdb.query("""SELECT 
             SPORT, MEDIAN(AGE) as median_age 
             FROM df_os GROUP BY SPORT 
             ORDER BY median_age DESC 
             LIMIT 5""").df()
```

```python
df_os.groupby("Sport")[["Sport", "Age"]].median(numeric_only=True).sort_values(
    by="Age", ascending=False
).head().reset_index().rename({"Age": "median_age"}, axis=1)
```

 f) Find out the five sports with lowest median age



```python
duckdb.query("SELECT SPORT, MEDIAN(AGE) as median_age FROM df_os GROUP BY SPORT ORDER BY median_age LIMIT 5").df()
```

  f) Find out top 10 countries after number of gold medals

```python
duckdb.query("SELECT NOC, COUNT(MEDAL) as num_medals  FROM df_os WHERE MEDAL = 'Gold' GROUP BY NOC ORDER BY num_medals DESC LIMIT 10").df()
```

 g) Find out top 10 countries after number of medals



```python
duckdb.query("SELECT NOC, COUNT(MEDAL) as num_medals  FROM df_os GROUP BY NOC ORDER BY num_medals DESC LIMIT 10").df()
```

  h) Plot a time series line chart of number of female and male atheletes in same graph.

```python
df_os.columns
```

```python
import seaborn as sns
```

```python
male_female = duckdb.query(
    """SELECT 
        YEAR, 
        SEX, 
        COUNT(SEX) as Total 
    FROM df_os 
    GROUP BY YEAR, SEX 
    ORDER BY YEAR"""
).df()
male_female
```

```python

sns.lineplot(male_female, x="Year", y="Total", hue="Sex")
```
